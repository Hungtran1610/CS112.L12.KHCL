# Tóm tắt

Tìm số ống nước CẦN để các nhà vẫn được kết nối và tổng những trường hợp tối ưu đó

# Nhận dạng

Đồ thị + toán hoán vị tổ hợp

# Thiết kế thuật toán
**_Input:_**
- Dòng đầu tiên chứa 2 số nguyên $N, M (1 \le N, M  \le 10^5)$
- Mỗi dòng trong M dòng tiếp theo chứa 2 số nguyên $a, b (1 \le a,  b \le N)$ cho biết vẫn còn đường ống nối nhà $a$ với nhà $b$

**_Output:_**
- Số đường ống cần được lắp thêm và số phương án đáp ứng yêu cầu(theo mô đun $10^9+7$)

**1. Ý TƯỞNG:**
- Giả sử ta có 7 nhà và sơ đồ đường ống nước chưa hư:
```
    3 === 4 === 1

    2 === 5

    6 === 7
```

- Q1: Cần ít nhất bao nhiêu đường ống nước để nối các nhà?
    * Mỗi dòng là mỗi dãy nhà có đường ống nước kết nối với nhau. Ta thấy sơ đồ dường như chia 7 nhà thành 3 nhóm nhà riêng biệt. Nhiệm vụ bây giờ là kết nối những nhóm nhà riêng biệt ở trên. Vậy để để 7 nhà vẫn được kết nối với nhau => cần tối thiểu 2 đường ống nước
 
- Q2: Có bao nhiêu trường hợp để lắp đặt những đường ống đó?
    * Vì các đường ống nối trực tiếp vào các nhà, nên ta chọn từng nhà trong từng các nhóm để xét. Vậy ta có xét tổ hợp các nhà trong từng nhóm như sau:

     * Nhóm 1 có $C_1$ = 3 cách chọn(nhà)
     * Nhóm 2 có $C_2$ = 2 cách chọn(nhà)         => theo nguyên tắc tổ hợp thì số cách lắp sẽ bằng `3x2x2`
     * Nhóm 3 có $C_3$ = 2 cách chọn(nhà)

    * Với 7 nhà với 2 đường ống thì ta có thể cố định một đường ống và hoán vị đường ống còn lại. Vì một nhà có thể liên kết tới 2 đường ống nên ta có $7$ ^ (2-1). Vậy với trường hợp trên, ta có thể tính ra số trường hợp là [ $7$ ^ (2-1)] $x$ 3 $x$ 2 $x$  2 = `84` trường hợp

Tổng quát hơn ta sẽ có công thức $T$ = [ $N$ ^ ($M$ - $1$) ] $*$ ($C_1 * C_2 * $ ... $ * C_N$), với:
- $T$ là số trường hợp.
- $N$ là số lượng nhà.
- $M$ là số lượng ống nước cần ít nhất.
- $C$ là cách số cách chọn trong từng nhóm nhà.

[ update 30/9 ] hoặc chúng ta quên đi những cái ở trên và đọc cái này https://vi.wikipedia.org/wiki/C%C3%A2y_bao_tr%C3%B9m

**2. THUẬT TOÁN:**
```
> B1: biểu diễn đồ thị.
> B2: cho ma trận đi qua thuật toán bfs/dfs ( hay cái gì đó ) để tìm ra các đỉnh được kết nối nhau
> B3: theo công thức mà tính số trường hợp
```

````python
n, m = map(int, input().split())
edges = {}
for _ in range(m):
    u, v = map(int, input().split())
    try:
        edges[u].append(v)
    except:
        edges[u] = [v]
    try:
        edges[v].append(u)
    except:
        edges[v] = [u]
      
      
visited = set()

def bfs(u):
    index = 0
    queue = [u]
    visited.add(u)  
    
    while index < len(queue):
        u = queue[index]
        index += 1
        if u not in edges:
            continue
        for v in edges[u]:
            if v not in visited:
                visited.add(v)
                queue.append(v)
        
    return len(queue)
    
ans = []
for i in range(1,n+1):
    if i not in visited:
        ans.append(bfs(i))
print(len(ans) - 1)
anw = 0

if len(ans) > 1:
    base = int(1e9 + 7)
    anw = 1
    anw2 = 1
    for _ in range(len(ans) - 2):
        anw2 = (anw2*n) % base
    for x in ans:
        anw = (anw * x) % base
    anw = anw * anw2 % base
    
print(anw)
````
