# Tóm tắt
Tìm số nguyên dương $m$ từ số $n$ sao cho $m$ chia hết cho 3 và khác $n$ đúng một chữ số

# Nhận dạng
Như theo đề bài, ta cần giải quyết 3 vấn đề:
- $m$ chia hết cho 3.
- $m$ chỉ khác $n$ ở một chữ số.
- $m$ lớn nhất có thể


# Thiết kế thuật toán
**1. Ý TƯỞNG:**

Thay đổi 1 chữ số của $n$ thành chữ số lớn nhất để tổng các chữ số trong $n$ chia hết cho 3
- Ta lần lượt tăng từng chữ số của $n$ lên 3 để nhận được số lớn nhất, nếu tìm được chữ số phù hợp thì dừng lại
- Nếu đến chữ số cuối cùng của n vẫn không thay đổi được, ta sẽ trừ $n$ một số $r$ - là số dư $r$ = $n$ % 3.

# Cài đặt

````python
n = int(input())

l = list(str(n))
x = 3 - n%3

check = True
i = 0

while check:
    p = int(l[i]) + x
    if (p < 10):
        while p < 7:
            p += 3
        l[i] = str(p)
        check = False
    i += 1
    
    if i == len(l):
        check = False

if "".join(l) != str(n):
    print("".join(l))
    
else:
    if n % 3 == 0:
        temp = int(l[-1]) - 3
    else:
        temp = int(l[-1]) - n%3
    l[-1] = str(temp)
    print("".join(l))
````
