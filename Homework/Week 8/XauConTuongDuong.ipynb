# Thiết kế thuật toán
**1. Ý TƯỞNG:**

Ta sử dụng hàm hash() trong python. Hàm hash() sẽ lấy nhận input là một nội dung bất kì (số nguyên, số thực, chuỗi,..) sau đó sử dụng các thuật toán để trả về một giá trị đầu ra có độ dài cụ thể.

Ví dụ ta sử dụng hash cho chuỗi "XauConTuongDuong", kết quả sẽ trả về giá trị là 4479018002602317167

**2. THUẬT TOÁN:**

Ta có $q$ truy vấn, thế nên thời gian tính toán có thể lên đến $O(n^2)$. Vì thế, ta tính tổng bằng cách công thức: Sum [$a$ –> $b$] =  PrefixSum[$b$] – PrefixSum[$a-1$]. Thời gian tính toán sẽ được giảm xuống thành $O(q+n)$ đến $O(n)$ và như thế đáp ứng yêu cầu đề bài.

**_Pseudo code:_**
```
Khởi tạo:
    array = [0] # mảng chứa tổng độ dài của các chuỗi con
Chương trình chính:
    1.Lặp i từ 0 tới len(string):
        Thêm array[vị trí trước đó] + giá trị hash()
    2.Sử dụng PrefixSum để tính tổng
    3.Lặp q lần truy vấn
        3.1 Kiểm tra 2 array [a,b] và [c,d] có bằng nhau không
        3.2 In ra YES nếu bằng và ngược lại NO.
In ra kết quả và kết thúc chương trình
```

# Cài đặt

````python
from sys import stdin, stdout

string = stdin.readline().strip()
q = int(input())

array = [0]

for word in string:
    array.append(array[-1] + hash(word))
    
def check(start, end):
    return array[end] - array[start-1]

for _ in range(q):
    a, b, c, d = map(int, stdin.readline().split())
    if check(a,b) == check(c,d):
        print("YES")
    else:
        print("NO")
````
