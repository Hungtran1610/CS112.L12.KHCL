# Tóm tắt

Tìm và đếm các cặp gene có chung cấu tạo giữa 2 gene.

# Nhận dạng

Dựa theo những dữ kiện mà ta có 2 gene kế nhau là một cặp, vậy bài toán đặt ra là tìm và các đếm cặp gene tìm được ở gene 1 và xác định số lần xuất hiện của chúng ở gene 2.

# Thiết kế thuật toán
**1. Ý TƯỞNG:**
- Với từng cặp gene trong gene 2, ta tăng giá trị ứng với cặp đó trong $gene pool$ lên 1.
- Với từng cặp gene trong gene 1, nếu cặp đó có trong $gene pool$, ta tăng biến $result$ lên 1.
- In kết quả là biến $result$.

**2. THUẬT TOÁN:**

**_Pseudo code:_**

```
init gene_1
init gene_2

init gene_pool = {}
init count = 0

for i from 1 to gene_2.lenght do
    if {gene_2[i], gene_2[i+1]} not in gene_pool do
        gene_pool add {gene_2[i], gene_2[i+1]}
    else do
        gene_pool[{gene_2[i], gene_2[i+1]}] = gene_pool[{gene_2[i], gene_2[i+1]}] + 1

for i from 1 to gene_1.lenght do
    if {gene_1[i]], gene_2[i+1]} in gene_pool do
        count++
```

# Cài đặt

````python
gene_1 = input().strip()
gene_2 = input().strip()

gene_pool = {}
for i in range(0, len(gen_2)-1):
    if (gene_2[i] + gene_2[i+1]) not in gene_pool:
        gene_pool[gene_2[i]+gene_2[i+1]] = 1
    else:
        gene_pool[gene_2[i]+gene_2[i+1]] += 1

res = 0
for i in range(0, len(gene_1)-1):
    if (gene_1[i] + gene_1[i+1]) in gene_pool:
        res += 1

print(res)
````
